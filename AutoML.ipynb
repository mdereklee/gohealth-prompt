```python
import h2o
from h2o.automl import H2OAutoML
h2o.init()
```

```python
train = h2o.import_file("train.csv")
train
```

```python
test = h2o.import_file("test.csv")
test
```

```python
target_encode_features = [
    'interview_date_mo',
    #'client_name_clean',
    'industry_clean',
    'location_clean',
    'position_to_be_closed_clean',
    'gender_clean',
    'current_location_clean',
    'job_location_clean',
    'interview_venue_clean',
    'marital_clean',
    'interview_type_clean',
    'candidate_native_location_clean',
    'skillset_clean',
    'q_start_time_clean',
    'q_unsch_mtgs_clean',
    'q_call_3hrs_clean',
    'q_alt_number_clean',
    'q_resume_clean',
    'q_venue_clean',
    'q_call_letter_clean',
    'q_total',
]

for f in target_encode_features:
    train[f] = train[f].asfactor()
    test[f] = test[f].asfactor()
    
response_col_name = 'attended_clean'
train[response_col_name] = train[response_col_name].asfactor()
test[response_col_name] = test[response_col_name].asfactor()
```

# AutoML

```python
aml_orig = H2OAutoML(max_runtime_secs=1200)
aml_orig.train(x=target_encode_features, y=response_col_name, training_frame=train, fold_column='fold')
```

```python
# Print Leaderboard (ranked by xval metrics)
# View the AutoML Leaderboard
lb_orig = aml_orig.leaderboard
lb_orig.head()  # lb.nrows)  # Print all rows instead of default (10 rows)
#lb.sort('aucpr', [False])
```

```python
model_ids = lb_orig.as_data_frame()['model_id'].tolist()
model_ids
```

```python
best_model_orig = h2o.get_model(model_ids[0])
```

```python
best_model_orig_perf = best_model_orig.model_performance(test)
best_model_orig_perf_train = best_model_orig.model_performance(train)
print('Accuracy(Train/Test):', best_model_orig_perf_train.accuracy()[0][0], '/', best_model_orig_perf.accuracy()[0][0])
print('AUC(Train/Test):', best_model_orig_perf_train.auc(), '/', best_model_orig_perf.auc())
```

```python
best_model_orig_perf_train.plot()
best_model_orig_perf.plot()
```

# Using target encoded variables

```python
aml_te = H2OAutoML(max_runtime_secs=1200)
x_ = [f'{c}_te' for c in target_encode_features]
aml_te.train(x=x_, y=response_col_name, training_frame=train, fold_column='fold')
```

```python
lb_te = aml_te.leaderboard
lb_te.head()
```

```python
model_ids_te = lb_te.as_data_frame()['model_id'].tolist()
model_ids_te
```

```python
# Get the "StackedEnsemble, BestOfFamily" Stacked Ensemble model
best_model_te = h2o.get_model(model_ids_te[0])
```

```python
best_model_te_perf = best_model_te.model_performance(test)
best_model_te_perf_train = best_model_te.model_performance(train)
print('Accuracy(Train/Test):', best_model_te_perf_train.accuracy()[0][0], '/', best_model_te_perf.accuracy()[0][0])
print('AUC(Train/Test):', best_model_te_perf_train.auc(), '/', best_model_te_perf.auc())
```

```python
best_model_te_perf_train.plot()
best_model_te_perf.plot()
```
